*Kapittel 3 - Bilisters tilpasning til bompenger*     

I dette kapitlet gjennomfrer du analysen av en bilist som en konsument som kan velge mellom to goder: "antall kilometer kjrt" og "andre goder". Prisen p andre goder kan du sette til 1 (dvs antall enheter av dette goder er et direkte ml p penger). "Andre goder" tegner konomer ofte p den vertikale aksen. Prisene pr kilometer for bilkjring for elbiler og biler med forbrenningsmotorer kan du finne p s. 41 i [dette dokumentet fra TI](https://www.toi.no/getfile.php?mmfileid=74283). Takstene for bompenger finner du hos [Bompengeselskap Nord](https://bpsnord.no/bypakke-tenk-tromso/) - dette er taksten per passering, s du m prve  gjre dette om til en kostnad per kilometer for bilister med ulik bilteknologi (si hvordan du har valgt  gjre dette).      

Du m selv velge hvilken nyttefunksjon du vil bruke til  gjengi konsumentens preferanser (vi har brukt 2 typer: Cobb-Douglas og Stone-Geary). Det er viktig at du begrunner valget av nyttefunksjonen!         

Din analyse br vise konsumentens tilpasning fr og etter priskningen, og du kan gjerne se p forskjellige typer bilister (de med elbil kontra de med fossilbil for eksempel). Vis substitusjons- og inntektseffektene, og bruk sistnevnte til  si noe om hvor mye bilistene taper p innfring av bompenger. (Er du virkelig ambisis kan du tenke p begrepene ["kompenserende/ekvivalente variasjonen"](https://www.sfu.ca/~wainwrig/Econ200/documents/cv-ev-notes.pdf) av en prisendring). Du kan ogs koble dette til begrepet konsumentoverskudd som for eksempel i kapittel 2.7 i [Ressursbruk i transportsektoren - noen mulige forbedringer](https://www.ntnu.no/documents/1261860271/1262010703/Concept_Nr44_no-ny.pdf/e6b903f5-b438-40f1-90d5-aeac5deb5472).  Legg merke til hva som brukes p aksene her!      

Strrelsen p effekten av en priskning henger sammen med ettersprselens priselastisitet (som ofte m mles indirekte for bilkjring). Estimat p priselastisiteter kan du finne [her](https://www.tiltak.no/b-endre-transportmiddelfordeling/b-1-styring-bilbruk/b-1-3/) (under Figur 1) og her i [Tabell 9](https://www.vtpi.org/elasticities.pdf) for eksempel.
