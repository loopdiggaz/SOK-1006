### 1. Tilpasning i godemarkedt på kort sikt
På kort sikt vil minst én produksjonsfaktor være fast, og dette vil utgjøre en del av bedriftens faste kostnader. Dersom bedriften produserer er produkt ved å kombinere kapital og arbeidskraft, og har en fast mengde med kapital på kort sikt ($K_0$) vil $rK_0$ være en del av bedriftens faste kostnad. Bedriften kan endre produksjon kun ved å endre mengde arbeidskraft som er brukt i produksjonsprosessen.        
Vi kan skrive fortjenesten som en funksjon av produsert mengde $F(x)$:

$$F(x)=px-(C_f+C_v(x)).\tag{1}$$

Vi kan se at fortjeneste er positiv dersom $px>C_f+C_v(x)$, dvs

$$F(x)>0 \Rightarrow p>\frac{C_f+C_v(x)}{x}=\overline{C}(x).\tag{2}$$

Prisen (gjennomsnittsinntekten) må være større enn gjennomsnittskostnaden for positiv fortjeneste. I tillegg til dette er vi ute etter verdien på x som maksmimerer $F(x)$. Førsteordensbetingelsen for et maksimum er

$$F'(x)=p-C'_v(x)=0\Rightarrow p=C'_v(x). \tag{3}$$

Med ord: prisen (her grenseinntekten) må være lik grensekostnaden for et maksimum. Dersom $p>GK$ kan bedriften øke fortjeneste ved å selge en ekstra enhet, og produksjon kan ikke ha vært optimal. Dersom $p<GK$ har bedriften tapt penger på den siste enheten solgt, og kunne ha redusert produksjon for å øke fortjeneste. Da kan ikke denne produksjonen ha vært optimal. Men forsteordensbetingelsen karakteriserer et ekstremt punkt, og vi bør sjekke at dette punktet er et maksimum. Vi gjør dette ved å se på den andre ordensbetingelsen for et maksimum:

$$F''(x)=-C''_v(x)<0\Rightarrow C''_v(x)>0. \tag{4}$$

Dette sier at grensekostnaden må være økende i maksimumspunktet. (*Forklaring i forelesningen!*).




#### 1.1 Eksempel
La oss illustrere med et eksempel (s. 173 i boka): $p=10, C_f=16, C_v=x^2$.
$$F(x)=10x-(16+x^2)$$
Førsteordensbetingelsen blir sa
$$F'(x)=10-2x = 0 \Rightarrow x=5$$
Andreordensbetingelse er oppfylt ettersom $F''(x)=-2 < 0.$ Da har vi funnet maksimal fortjeneste $F(5) = 10(5)-(16+5^2)=9$.


```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def fortj(p, x):
    return p*x-(16+x**2)
```

### 2. Fra produktfunksjon til tilbudsfunksjon

For å produsere må bedriften bruke innsatsfaktorer. La oss anta at disse er Arbeidskraft (N) og Kapital (K), og at kapitalmengden er fast på kort sikt $K=K_0$. Produktfunksjonen på kort sikt er dermed
$$x=f(N,K_0)\tag{5}$$

Vi løser (5) for $N$ slik at vi kan skrive $N(x)$ som angir hvor mye arbeidskraft som må ansettes for å produsere $x$ enheter vare (dette er også en funksjon av $K_0$). Da kan vi skrive kostnadsfunksjonen som
$$ C(x) = rK_0 + wN(x)\tag{6}$$

Da har vi kostnadsfunksjonen på samme form som vi brukte i (1), og kan finne bedriftens tilbud!

#### 2.1 Eksempel
En bedrift produserer en vare med produktfunksjon $x = N^{\frac{2}{5}}K^{\frac{1}{4}}$; videre setter vi $K_0=1, w=3, r=5$. Hvor mye bør bedriften produsere for å maksimere fortjeneste på kort sikt?       
Her kan vi ikke sette opp (1) ovenfor ettersom vi har informasjon om produktfunksjonen, men ikke kostnadsfunksjonen. Dette kan vi få:

```python
import sympy as sp
x,N,p = sp.symbols('x N p', positive=True, real=True) 
```

```python
def prod_ks(N):
    return N**.4

# løs x=f(N)
N_ks = sp.solve(prod_ks(N)-x,N)[0]
N_ks
```

Vi har at $N=x^{\frac{5}{2}}$. Dette kan vi bruke for å finne kostnadsfunksjonen:

```python
Kostnad=5+3*N_ks
Kostnad
```

```python
# tilbudet gis av likning (3) p=GK
GK = sp.diff(Kostnad, x)
GK
```

Tilbudsfunksjonen er $p=7.5x^{\frac{3}{2}}$, gitt at $p>\overline{C}(x)$ (positiv fortjeneste) og $C''(x)>0$ (andre ordensbetingelse).

```python
AOB=sp.diff(GK,x)
AOB
```

```python
# vi gjør klart for å plotte
Kostnad_fn=sp.lambdify(x, Kostnad)

GK_fn=sp.lambdify(x, GK)

display(Kostnad_fn(x))
display(GK_fn(x))
```

```python
def AC(x):
    return Kostnad_fn(x)/x
# finn lavest x som gir positiv fortjeneste, dvs reservasjonsprisen

lavest_x=sp.solve(AC(x)-GK_fn(x),x)[0]
lavest_x
```

```python
x_num=np.linspace(0.01,10,100)
x_1=np.linspace(float(lavest_x),10,100)
fig, ax = plt.subplots()
ax.set_ylabel('Pris', loc='top')
ax.set_xlabel('Kvantum (x)', loc='right')
ax.set(xlim=(0,2))
ax.set(ylim=(0,20))
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

ax.plot(x_num, GK_fn(x_num), label = 'GK')
ax.plot(x_1, GK_fn(x_1), color='red', label = 'Tilbud (p=GK)')
ax.plot(x_num, AC(x_num), label = 'Gjennomsnittskostnad')


ax.set_title('Bedriftens tilbud')
ax.legend(loc='best');
```

Tilbudsfunksjonen er den delen av grensekostnadskurven som ligger over gjennomsnittskostnad.

### 3. Bedriftens tilpasning i faktormarkedet på kort sikt

La oss anta at bedriften produserer en vare av Arbeidskraft (N) og Kapital (K), og at kapitalmengden er fast på kort sikt $K=K_0$. Vi kan nå se på hva det er som bestemmer hvor mye arbeidskraft bedriften vil etterspørre for å maksimere fortjeneste på kort sikt.                  
         

1. Sett opp fortjenestesfunksjonen på kort sikt:     
$$ F(N) = p f(N, K_0) - (rK_0 + w N)\tag{7}$$

hvor $f(N, K_0)$ er produktfunksjonen på kort sikt, $rK_0$ er faste kostnader og $wN$ er variable kostnader.         

2. Sett opp førsteordensbetingelsen:
$$ F'(N) = p f'_N(N, K_0) - w = 0\Rightarrow p f'_N(N, K_0) = w  \tag{8}$$

Venstresiden er *verdien av grenseproduktet* som den siste enheten med arbeidskraft produserer, og høyresiden er *grensekostnaden* til den siste enheten med arbeidskraft.        

3. Løsningen til (8) er $N(K_0, w, p)$, dvs optimal mengde arbeidskraft etterspurt på kort sikt. Legg merke til at dette ikke avhenger av $r$ (hvorfor ikke?).          

#### Eksempel 3.1
Vi antar samme produktfunksjon som før $x = N^{\frac{2}{5}}K^{\frac{1}{4}}$, og setter opp problemet på generelt vis med $k$ som fast mengde kapital på kort sikt.


```python
# Overskuddsmaksimering på kort sikt

k,w,r = sp.symbols('k w r', real=True, positive=True)

def prod(N,k):
    return N**(0.4)*k**(0.25)
def profitt_ks(N, k, p, w, r):
    return p*prod(N,k) - w*N-r*k
profitt_ks(N,k, p, w, r)
```

```python
# maksimer mhp N
dpi_N=sp.diff(profitt_ks(N,k, p, w, r),N)
foc_N=sp.Eq(dpi_N,0)
foc_N
```

```python
# Løs for N
N_ks=sp.solve(foc_N,N)[0]
N_ks
```

```python
# sjekk andreordensbetingelse for et maksimum
sp.diff(dpi_N,N)
```

```python
N_sol=N_ks.subs([(p,10),(k,1)])
N_sol
```

```python
# Vi plotter FOB for k=1

def mrp(N,k, p):
    return (.4*p*k**.25)/N**.6

n=np.linspace(0.1,10,100)



fig, ax = plt.subplots()
ax.set_ylabel('w (NOK)', loc='top')
ax.set_xlabel('Arbeidskraft (N)', loc='right')
ax.set(xlim=(0,10))
ax.set(ylim=(0,10))
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

ax.plot(n, mrp(n, 1, 10), label = 'Etterspørsel etter N (p=10)')

ax.hlines(3,0,10, color= 'red', label = 'w=3')
ax.hlines(5,0,10, color= 'gray', label = 'w=5')
ax.vlines(float(N_sol.subs(w,5)),0,5, linestyle='dashed')
ax.vlines(float(N_sol.subs(w,3)),0,3, linestyle='dashed')

ax.set_title('Etterspørsel etter arbeidskraft (k=1)')
ax.legend(loc='best');
```

Vi ser at bedriften etterspør mindre arbeidskraft når lønna går opp. Dette er i tråd med andre etterspørselsfunksjoner som vi kjenner til.        

Hva skjer når produktprisen øker?
