```python
import pandas as pd

df=pd.read_csv("https://uit-sok-1006-v23.github.io/data/suits_watermelons.csv",delimiter=";")

print(df.head())
```

```python
# 7.1
# 1. 
# a) 

import numpy as np

# Lager ny kollone med P og Q. 

df['P'] = np.exp(df['log p'])
df['Q'] = np.exp(df['log q'])
df['H'] = np.exp(df['log h'])
df
```

```python
# b) 

from matplotlib import pyplot as plt

# Bruker axs for å definere aksene
fig, axs = plt.subplots(1,2, figsize=(10,5))

# Plott
axs[0].plot(df['Year'],df['Q'], color = 'black', linestyle ='dashed', label = 'crop', marker = 'o', linewidth=0.5)
axs[0].plot(df['Year'],df['H'], color = 'black', label = 'harvest')
axs[1].plot(df['Year'],df['P'], color = 'black', label = 'price', marker = 'o', linewidth=0.5)

# Legend
axs[0].legend(loc='best', frameon = False)
axs[1].legend(loc='best', frameon = False)

# Label
axs[0].set_ylabel('Price (p)', loc='top')
axs[0].set_xlabel('Year', loc='right')
axs[1].set_ylabel('Price (p)', loc='top')
axs[1].set_xlabel('Year', loc='right')

# Tittel 
axs[1].title.set_text("Price p (dollars per thousand)") 
axs[0].title.set_text("Crop q and harvest h (millions)")
```

```python
# 7.2 
# Plot demand og supply curves.

# 2. 
# a)

import pandas as pd

q=np.linspace(20,100,17)

df2 = pd.DataFrame(q)

# b) 

log_q = np.log(q)

log_q = np.array(log_q)

log_price = 8.5 - 0.82*log_q

log_price_s = -2 + 1.7*log_q 

# c)

supply = np.exp(log_price_s)

demand = np.exp(log_price)

df2 = pd.DataFrame(list(zip(q, log_q, log_price_s, log_price, supply, demand)),
                columns =['Q', 'Log q', 'supply (log p)','demand (log p)','supply','demand'])
df2 
```

```python
# d) 

fig,ax=plt.subplots()
ax.plot(df2['Q'],df2['supply'], color = 'red', label = 'supply')
ax.plot(df2['Q'],df2['demand'], color = 'blue', label = 'demand')
ax.set_ylabel('Pris (p) per tusen', loc='top')
ax.set_xlabel('Kvantum (x) millioner', loc='right')
ax.legend(loc='best')
ax.set_title('Vannmelon (tilbud -og etterspørsel)')
```

```python
# 3. 
# a)

shock = -2.0 + 1.7*log_q + 0.4

new_supply = np.exp(shock)

df2 = pd.DataFrame(list(zip(q, log_q, log_price_s, log_price, supply, demand, shock, new_supply)),
                columns =['Q', 'Log q', 'supply (log p)','demand (log p)','supply','demand', 'new supply (log p)', 'new supply'])
df2
```

```python
#  b) 
fig, axs=plt.subplots()
axs.plot(df2['Q'],df2['supply'], color = 'red', label = 'supply')
axs.plot(df2['Q'],df2['demand'], color = 'blue', label = 'demand')
axs.plot(df2['Q'],df2['new supply'], color = 'green', label = 'new supply (shock)')
axs.set_ylabel('Pris (p) per tusen', loc='top')
axs.set_xlabel('Kvantum (x) millioner', loc='right')
axs.legend(loc='best')
axs.set_title('Vannmelon (tilbud -og etterspørsel)')
```

```python
print(f'Pris reflekterer knapphet, om et gode blir dyrere å produsere antok jeg at prisen ville stige og tilbudet falle.') 
print(f'Tilbudet har fått et venstre skift, og prisen presses opp til ny likevekt, dette passer godt med hypotesen min.') 
```

```python
7.2 

import sympy as sp
from sympy import symbols

# a) 

log_p, log_q  = symbols('log_p, log_q') # Lager symboler

eq = sp.Eq(-2 + 1.7*log_q,log_p) # Løser ligning. 

display(sp.solve(eq,log_q))

print(f'I den nye ligningen ser vi at tilbudet er uelastisk og relativt lite sensitiv for prisendringer.') 
print(f'Når prisen øker med 1 prosent, øker tilbudet med 0.58 prosent.')
print(f'Fordelen med å bruke log er at vi kan se elastisten utfra konstanten som ganges med variabelen log p.')
```

```python
# b) 

eq2 = sp.Eq(8.5 - 0.82*log_q,log_p) # Løser ligningen.

display(sp.solve(eq2,log_q))

print(f'I den nye ligningen ser vi at etterspørselen er elastisk og dermed sensitiv for prisendringer.') 
print(f'Når prisen øker med 1 prosent, faller etterspørselen med 1.21 prosent.')
```
