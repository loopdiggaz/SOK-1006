```python
import numpy as np
import sympy as sp

N, A, K = sp.symbols('N A K', positive=True, real=True)

def prod(N,K):
    produktfunksjon = 100*N*K
    return produktfunksjon

prod(N,K)
```

```python
gp2 = sp.diff(prod(N,2),N)

def gp(K):
    grenseprodukt = 100*K
    return grenseprodukt

gp(2)

gp2
```

Av utrykket fremgår det at dersom K = 2 vil grenseproduktet være = 200. Det betyr at dersom vi øker bruken av arbeidskraft med en enhet øker produksjonen med 200 enheter.


```python
def ap(N,K):
    gjennomsnitt = prod(N,K)/N
    return gjennomsnitt

ap(N,K)
```

Merk at med denne produktfunksjonen er gjennomsnittsproduktiviten og grenseproduktivtiten like store. Dette skydes at produktfunksjonen er lineær.

```python
gp2_fn = sp.lambdify(N,gp2)
```

```python
from matplotlib import pyplot as plt

n = np.linspace(1,10,100)

fig, ax = plt.subplots()
ax.plot(n, prod(n,2), label = "Produktfunksjon")
ax.plot(n, gp(n), label = "Grenseprod")
ax.plot(n,ap(0,n), label = "Gjennomsnittsprod")

ax.legend(loc='best', frameon = False);
```
